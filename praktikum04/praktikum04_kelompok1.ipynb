```python
import matplotlib.pyplot as plt
# data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]
# membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
# memberi judul pada grafik
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.show()
```

```python
import matplotlib.pyplot as plt
# data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Science'
]

jumlah_mahasiswa = [120, 150, 100, 80]
# membuat basic bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
# menambah label pada bar
for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v+5, str(v), ha='center', va='bottom')
# memberi judul pada grafik
plt.title('Jumlah Mahasiswa Per Jurusan')
plt.show()
```
