```python
import matplotlib.pyplot as plt
```

## **Dataset**

```python
# Buat data sample
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
product_A_sales = [10,20,15,25,30,45,40,50,60,55,65,70,]
product_B_sales = [5,10,8,15,18,20,22,30,25,35,40,45]
```

## **Single Line Plot**

```python
# line plot satu garis, gaya grafik default/standar
plt.plot(months, product_A_sales, label='product A')
plt.title('Penjualan Produk A per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

## **Multiple Line Plot & Customizations**

```python
# Kustomisasi line plot (line 1 - product A)
plt.plot(months, product_A_sales, label='Product A', linestyle="--", marker='o', color='Blue')
# Kustomisasi line plot (line 1 - product B)
plt.plot(months, product_B_sales, label='Product B', linestyle="-", marker='x', color='Red')

plt.title('Penjualan Produk A dan B per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

## **Jenis Garis untuk Menunjukkan Tren**

```python
# data sample tamabahn
product_C_sales = [18,2,25,67,45,89,45,32,86,80,65,45]
product_D_sales = [7,6,16,9,21,30,29,17,5,15,18,22]
# plot product A
plt.plot(months, product_A_sales, label='Product A', linestyle="--", color='Blue')
# Kustomisasi line plot product B (line 2 - product B)
plt.plot(months, product_B_sales, label='Product B', linestyle="-", color='Red')
# plot product C
plt.plot(months, product_C_sales, label='Product C', linestyle=":", color='Green')
# Kustomisasi line plot product D (line 4 - product D)
plt.plot(months, product_D_sales, label='Product D', linestyle="-.", color='Purple')

plt.title('Penjualan Produk per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.legend()
plt.grid(True)
plt.show()
```

## **Subplot pada Line Chart**

```python
# subplot 2 baris dengan 1 kolom, panjang 10 inch dan tinggi 8 inch
fig, axs = plt.subplots(2, 1, figsize=(10,8))

# plot pertama untuk product C
axs[0].plot(months, product_C_sales, label='Product C', color='green', marker='d')
axs[0].set_title('Penjualan Produk C per Bulan')
axs[0].set_xlabel('Bulan')
axs[0].set_ylabel('Jumlah Penjualan')
axs[0].legend()
axs[0].grid(True)

# plot pertama untuk product D
axs[1].plot(months, product_D_sales, label='Product D', color='purple', marker='s')
axs[1].set_title('Penjualan Produk D per Bulan')
axs[1].set_xlabel('Bulan')
axs[1].set_ylabel('Jumlah Penjualan')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
